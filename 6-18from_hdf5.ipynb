```python
import pandas as pd

# Load the DataFrame from the CSV file
df_results = pd.read_csv('df_results.csv', parse_dates=['time'])

# Initialize an empty list to hold the new rows
new_rows = []

# Function to aggregate values for a given period
def aggregate_values(start_time, end_time, label_time):
    period_df = df_results[(df_results['time'] > start_time) & (df_results['time'] <= end_time)]
    aggregated_row = period_df.drop(columns=['time']).sum().round(2)  # Round values to two decimal places
    aggregated_row['time'] = label_time
    new_rows.append(aggregated_row)

# Get the unique dates in the DataFrame
unique_dates = df_results['time'].dt.date.unique()

for date in unique_dates:
    # Define time periods for the current date
    date = pd.Timestamp(date)
    start_time_1 = date + pd.Timedelta(hours=18, minutes=30)
    end_time_1 = (date + pd.Timedelta(days=1)).replace(hour=6, minute=0, second=0)
    label_time_1 = end_time_1
    
    start_time_2 = date + pd.Timedelta(hours=6, minutes=30)
    end_time_2 = date.replace(hour=18, minute=0, second=0)
    label_time_2 = end_time_2
    
    # Aggregate values for the night period (18:30 to 06:00 next day)
    aggregate_values(start_time_1, end_time_1, label_time_1)
    
    # Aggregate values for the day period (06:30 to 18:00)
    aggregate_values(start_time_2, end_time_2, label_time_2)

# Convert the new rows to a DataFrame
new_df = pd.DataFrame(new_rows)

# Ensure 'time' column is the first column
new_df = new_df[['time'] + [col for col in new_df.columns if col != 'time']]

# Sort the DataFrame by 'time'
new_df = new_df.sort_values(by='time').reset_index(drop=True)

# Save the new DataFrame to a CSV file
new_df.to_csv('6-18_04.2021.csv', index=False)

print(new_df)
```
